```python
import pandas as pd
```

```python
train = pd.read_csv(r"F:\Desktop\DataSets\Loan_Prediction\train.csv")
test = pd.read_csv(r"F:\Desktop\DataSets\Loan_Prediction\test.csv")
```

```python
train.head()
```

```python
test.head()
```

```python
train = train.set_index('Loan_ID')
test = test.set_index('Loan_ID')
```

```python
print(train.isnull().sum(), train.info())
```

```python
print(test.isnull().sum(), test.info())
```

```python
train['Loan_Status'] = train['Loan_Status'].astype('category')
columns = ['Gender', 'Married', 'Dependents', 'Self_Employed', 'Education', 'Property_Area']
for col in columns:
    train[col] = train[col].astype('category')
    test[col] = test[col].astype('category')
```

```python
print(train.isnull().sum(),test.isnull().sum() )
```

```python
cat_misscols = columns[:-2]
```

```python
cat_misscols
```

```python
for cols in cat_misscols:
    mode = train[cols].value_counts().index[0]
    train[cols] = train[cols].fillna(mode)
```

```python
train.isnull().sum()
```

```python
for cols in cat_misscols:
    test_mode = test[cols].value_counts().index[0]
    test[cols] = test[cols].fillna(test_mode)
```

```python
print(test.isnull().sum(), test.info())
```

```python
num_misscols = ['LoanAmount', 'Loan_Amount_Term', 'Credit_History']

from sklearn.preprocessing import Imputer
imp = Imputer(missing_values = 'NaN', strategy = 'median', copy = False)
for col in num_misscols:
    train[col] = imp.fit_transform(train[col].values.reshape(-1, 1))
    test[col] = imp.fit_transform(test[col].values.reshape(-1, 1))
```

```python
print(train.info(), train.isnull().sum())
```

```python
print(test.isnull().sum(), test.info())
```

```python
train.head()
```

```python
dum_cols = columns
```

```python
dum_cols
```

```python
train1 = pd.get_dummies(train, columns = dum_cols, drop_first = True)
test1 = pd.get_dummies(test, columns = dum_cols, drop_first = True)
```

```python
train1.info()
```

```python
train1.head()
```

```python
test1.head()
```

```python
x = train1.drop(['Loan_Status'], axis = 'columns')
y = train1['Loan_Status']
print(x.shape, y.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors = 5, weights = 'distance')
model.fit(x, y)
```

```python
train_preds = model.predict(x)
test_preds = model.predict(test1)
print(train_preds.shape, test_preds.shape)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
cf = confusion_matrix(y, train_preds)
clr = classification_report(y, train_preds)
ac = accuracy_score(y, train_preds)
print(cf)
print(clr)
print('accuracy:', ac * 100)
```

```python
submission = pd.DataFrame(test_preds, index = test1.index, columns = ['Loan_Status'])
submission.head()
```

```python
submission.to_csv('KNNLoan_Predictor.csv', index = True)
```
